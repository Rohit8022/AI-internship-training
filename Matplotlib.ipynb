```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('sharma-kohli.csv')
data
```

# LINE CHART

```python
# All the parameters availabe with the plots
plt.plot(data['index'],data['V Kohli'],color='red',marker='*',linewidth=1,markersize=10,linestyle='-.')
plt.plot(data['index'],data['RG Sharma'],color='blue',marker='o',linewidth='2',markersize=2,linestyle='--')
plt.xlabel('Seasons')
plt.ylabel('Total RUna')
plt.title('Rohit VS Virat runs in each seasons')
```

```python
# legend in plot 
plt.plot(data['index'],data['V Kohli'],color='red',marker='*',linewidth=1,markersize=10,linestyle='-.',label='VK')
plt.plot(data['index'],data['RG Sharma'],color='blue',marker='o',linewidth='2',markersize=10,linestyle=':',label='RS')
plt.xlabel('Seasons')
plt.ylabel('Total RUna')
plt.title('Rohit VS Virat runs in each seasons')
plt.legend() # LOCATION OF LEGEND CAN ALSO BE MANUALLY SELECTED
```

```python
# limiting axis
price=[35000,30000,40000,45000,37000,50000,61000,55000,70000,450000]
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
plt.plot(year,price) # in this case the last value effecte the chat for all previous values 
plt.ylim(20000,70000) # can set limit to see all the value bygtriminf the chart 
plt.xlim(2011,2015) 
```

```python
# Grid
price=[35000,30000,40000,45000,37000,50000,61000,55000,70000,450000]
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.grid() #to see the pts clearly properly 
```
