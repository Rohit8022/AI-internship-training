```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('sharma-kohli.csv')
data
```

# LINE CHART

```python
# All the parameters availabe with the plots
plt.plot(data['index'],data['V Kohli'],color='red',marker='*',linewidth=1,markersize=10,linestyle='-.')
plt.plot(data['index'],data['RG Sharma'],color='blue',marker='o',linewidth='2',markersize=2,linestyle='--')
plt.xlabel('Seasons')
plt.ylabel('Total RUna')
plt.title('Rohit VS Virat runs in each seasons')
```

```python
# legend in plot 
plt.plot(data['index'],data['V Kohli'],color='red',marker='*',linewidth=1,markersize=10,linestyle='-.',label='VK')
plt.plot(data['index'],data['RG Sharma'],color='blue',marker='o',linewidth='2',markersize=10,linestyle=':',label='RS')
plt.xlabel('Seasons')
plt.ylabel('Total RUna')
plt.title('Rohit VS Virat runs in each seasons')
plt.legend() # LOCATION OF LEGEND CAN ALSO BE MANUALLY SELECTED
```

```python
# limiting axis
price=[35000,30000,40000,45000,37000,50000,61000,55000,70000,450000]
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
plt.plot(year,price) # in this case the last value effecte the chat for all previous values 
plt.ylim(20000,70000) # can set limit to see all the value bygtriminf the chart 
plt.xlim(2011,2015) 
```

```python
# Grid
price=[35000,30000,40000,45000,37000,50000,61000,55000,70000,450000]
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.grid() #to see the pts clearly properly 
```

# BAR CHART

```python
# simple making
like=[10,20,40,12,30]
color=['red','yellow','green','blue','pink']
plt.bar(color,like)
```

```python
# horizontal bar
plt.barh(color,like)
```

```python
# Multi BarChart
df= pd.read_csv(r"C:\Users\vedeh\OneDrive\Desktop\AI_internship\batsman_season_record.csv")
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
```

```python
# Problen if the X axis name in too big
like=[10,20,40,12,30]
color=['red red red red fed red','yellowwwwwwwwwwwwwww','greeeeeeeeeeeen','blllllllllue','pppppppppppppppppink']
plt.bar(color,like)
plt.xticks(rotation='vertical') # or plt.xticks(rotation=90)
# Used to rotatle the name of x axis 
```

```python
# Stacked bar chart
plt.bar(df['batsman'],df['2017'])
plt.bar(df['batsman'],df['2016'],bottom=df['2017'])
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']))
```

# PIE CHART

```python
# simple data
data=[20,50,41,20,30]
plt.pie(data)
plt.show()
```

```python
# ON data set
df=pd.read_csv(r'C:\Users\vedeh\OneDrive\Desktop\AI_internship\Date_sets\gayle-175.csv')
df
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
# extract a pie
plt.pie(df['batsman_runs'],labels=df['batsman'],explode=[0.2,0,0,0,0,0])# can increase the distance by increasingb the value like 0.3 or 0.2
plt.show()
```

```python
# shadow in pie chat
plt.pie(df['batsman_runs'],labels=df['batsman'],shadow=True)
plt.show()
```
