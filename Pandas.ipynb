#  The sort_values() method

```python
# Applied on both series and DF
# on series->
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values(ascending=False).head() # head to get top 5
# ascending=False means DESC and ascending=True means ASC
```

```python
# on DF->
data.sort_values('city',ascending=False) #(col_name,ascending=T/F,Inplace=T/F)
# default ascending=True
# default inplace= False
#Inplace = it is done to make the chnages in the DF permanently by selecting it as True
```

```python
data.sort_values(['city','date']) # can also be applied on multi col by using a list
```

```python
# can also select one col in desc and other in asc
data.sort_values(['city','date'],ascending=[False,True])
```

## The Groupby() function

```python
# this method is used to group data based on one or more columns and then perform operations 
# like sum, count, mean, min, max, etc. on each group.
seasons=data.groupby('season')
```

```python
# to know in a season which was the first match and the last
seasons.first()
seasons.last()
len(data)      # To know the no. of rows in the DF
len(seasons)
```

```python
# to know the no. of rows/matches in each seasons
seasons.size().sort_values(ascending=False)
```

```python
# it returns a dictonary where key ia season and the value is a list of all matches if the season
seasons.groups # ATTRIBUTR
```

```python
seasons.get_group(2009).shape # to get deatils of all elements in a group if the grouped object is string use ('name')
```

```python
# agg functin will apply only if the valuse are numaric
seasons.sum() # can be done on specific group like 
               #seasons[2008].mean()
```

# deliveries.csv data set analysis Unsing GROUP BY

```python
df=pd.read_csv('deliveries.csv')
df
```

```python
runs=df.groupby('batsman')
```

```python
# TOP 5 BETSMAN WITH THE HIGHEST RUNS IN IPL
runs['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
# who hited the most 6's in IPL
fours=df[df['batsman_runs'] == 6]
fours.shape
```

```python
fours.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head()
```

```python
# highest run agaunsh team 
kohli=df[df['batsman']== 'V Kohli']
kohli.groupby('bowling_team')['total_runs'].sum().sort_values(ascending=False).head()
```

```python
# can also make a function for every batsman
def  hig_runs(batsman):
    kohli=df[df['batsman']== batsman]
    return kohli.groupby('bowling_team')['total_runs'].sum().sort_values(ascending=False).head() 
hig_runs('CH Gayle')    
```
