```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'C:\Users\vedeh\OneDrive\Desktop\AI_internship\Date_sets\mtcars.csv')
data.head()
```

```python
# FOR MAKINF BAR CHART
sns.barplot(x='cyl',y='mpg',data=data)
```

```python
## FOR MAKING OF COUNT CHART
sns.countplot(x='cyl',data=data,palette='Set2') #palette is used in categorical plots where colors represent categories.
plt.show()
```

```python
# HORIZONTAL COUNTCHAR
sns.countplot(y='cyl',data=data,palette='Set2')
```

```python
sns.countplot(x='cyl',hue='gear',data=data,palette='Set2')#hue is used to add one more categorical variable to a plot using different colors.
```

```python
# DISTRIBUTION PLOT
sns.displot(data=data,x='mpg',bins=10)
```

```python
# HEAT MAP
sns.heatmap(data.select_dtypes(include='number').corr(),annot=True,linewidths=0.5)
# add this ".select_dtypes(include='number')" because heatmap do not works on strings in this
#  data the the model is string 
```
